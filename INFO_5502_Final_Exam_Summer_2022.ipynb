<span style='background:yellow'>**Question 2 (100 pt).**</span> Write a python program to evaluate the text classification performance using accuracy, precision, recall, and f1 score. 

Accuracy = (TP + TN) / (TP+ TN+ FP+ FN)

Recall = (TP) / (TP+FN)

Precision = (TP) / (TP+FP)

F1 Score = (2*Precision*Recall) / (Precision + Recall)

Here is the confusion matrix of a text classfication task (six categories) on the testing data. 


```python
#### You code here

def metrics(actual_value, predicted_value):

  if actual_value == 1 and predicted_value == 1:
    true_positive = sum(actual_value) + sum(predicted_value)
  
  if actual_value == 0 and predicted_value == 0:
    true_negative = sum(actual_value) + sum(predicted_value)

  if actual_value == 1 and predicted_value == 0:
    false_negative = sum(actual_value) + sum(predicted_value)

  if actual_value == 0 and predicted_value == 1:
    false_positive = sum(actual_value) + sum(predicted_value)
    
  return true_positive, true_negative, false_negative, false_positive


def accuracy(true_positive, true_negative, false_negative, false_positive):
  return ((true_positive + true_negative) * 100) / true_positive + true_negative + false_negative + false_positive


def precision(true_positive, false_positive):
  return(true_positive * 100) / true_positive + false_positive


def recall(true_positive, false_negative):
  return (true_positive * 100) / true_positive + false_negative


def f1_score(true_positive, true_negative, false_negative, false_positive):
  prec = precision(true_positive, false_positive)/100
  recll = recall(true_positive, false_negative)/100
  f1_score = (2*prec*recll)/prec+recll
  return f1_score
```
