```python
# You code here (Please add comments in the code):

# importing necessary modules
import pandas as pd              # pandas for dataframe
import requests                  # requests to send http requests to amazon
from bs4 import BeautifulSoup    # BeautifulSoup to parse the html
from datetime import datetime    # datetime to save the csv file as a date

# headers to bypass and be able to scrape amazon
# found by inspecting page, going to Network, copying first file as cURL (bash), transforming to python format 
headers = {
    'authority': 'www.amazon.com',
    'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
    'accept-language': 'en-GB,en;q=0.9,en-US;q=0.8,fr;q=0.7,es-419;q=0.6,es;q=0.5',
    'cache-control': 'max-age=0',
    # Requests sorts cookies= alphabetically
    # 'cookie': 'session-id=145-3052181-6604163; i18n-prefs=USD; skin=noskin; ubid-main=134-5698716-5974708; lc-main=en_US; session-id-time=2082787201l; av-timezone=America/Chicago; session-token=/y+u5JOruQhjiyUlmvMudY5K2tXmf/Mo1k9tpdk9PksSwxgzq5bugoj/hGKCDpOIIP0ILoZ47BWG43FAPi4WFjULlHISWR0GrucaLEWuwmr78sDMms6lR/XSobawnJnEN5QHe9ml2B9+HC4kvRfodHXMOOpX2oqrOVueaIPNYfoL41boHwbkk0mDqW0jFG2U3sE42MJDqIgOX083YC006w==; csm-hit=tb:NZTZ3SQK18MQGJ30526E+sa-6JGP1NRME3SY63NZJ8WX-GP9RMEK7667BZ5DKEZA6|1658844585175&t:1658844585175&adb:adblk_yes',
    'device-memory': '8',
    'downlink': '10',
    'dpr': '1.25',
    'ect': '4g',
    'referer': 'https://www.amazon.com/Live-Die-Repeat-Edge-Tomorrow/dp/B00KW27K4K/ref=sr_1_1_sspa?crid=3P1IXE0AIU7DN&keywords=movies&qid=1658844509&sprefix=movie%2Caps%2C171&sr=8-1-spons&psc=1&spLa=ZW5jcnlwdGVkUXVhbGlmaWVyPUEyWERZMERIODFENFomZW5jcnlwdGVkSWQ9QTEwNDI1MDRUSUQzQzMwQkZOTU8mZW5jcnlwdGVkQWRJZD1BMDU5ODE2MjJCOFcyUE1JVU9TRU4md2lkZ2V0TmFtZT1zcF9hdGYmYWN0aW9uPWNsaWNrUmVkaXJlY3QmZG9Ob3RMb2dDbGljaz10cnVl',
    'rtt': '100',
    'sec-ch-device-memory': '8',
    'sec-ch-dpr': '1.25',
    'sec-ch-ua': '".Not/A)Brand";v="99", "Google Chrome";v="103", "Chromium";v="103"',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '"Windows"',
    'sec-ch-viewport-width': '1229',
    'sec-fetch-dest': 'document',
    'sec-fetch-mode': 'navigate',
    'sec-fetch-site': 'same-origin',
    'sec-fetch-user': '?1',
    'upgrade-insecure-requests': '1',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36',
    'viewport-width': '1229',
}


# getting the first page of urls for amazon.com and amazon.co.uk
urls_us = ["https://www.amazon.com/product-reviews/B00KW27K4K/ref=cm_cr_arp_d_viewopt_srt?ie=UTF8&reviewerType=all_reviews&sortBy=recent&pageNumber=1"]
urls_uk = ["https://www.amazon.co.uk/product-reviews/B00O0NZ6IY/ref=cm_cr_arp_d_viewopt_srt?ie=UTF8&reviewerType=all_reviews&sortBy=recent&pageNumber=1"]


# getting the next pages by adding each number from 2 to 100 at the end of the url
for n in range(2, 101):
  link_us = "https://www.amazon.com/product-reviews/B00KW27K4K/ref=cm_cr_arp_d_viewopt_srt?ie=UTF8&reviewerType=all_reviews&sortBy=recent&pageNumber="
  next_page_us = link_us + str(n) # adding each number to the string
  urls_us.append(next_page_us)    # appending each url to the list

  link_uk = "https://www.amazon.co.uk/product-reviews/B00O0NZ6IY/ref=cm_cr_arp_d_viewopt_srt?ie=UTF8&reviewerType=all_reviews&sortBy=recent&pageNumber="
  next_page_uk = link_uk + str(n)
  urls_uk.append(next_page_uk)


# defining a function with 1 parameter for a soup object
def get_usernames(a_soup):          
  try:
    usernames = a_soup.find("div", {"class": "a-profile-content"}).get_text() # finding specific html for usernames, only getting the text
    return usernames.strip() # removing leading and trailing spaces
  except Exception as e:
    print(e)


# defining other functions the same way for all other information wanted starting with the numbers of stars
def get_stars(a_soup):
  try:
    stars = a_soup.find("span", {"class": "a-icon-alt"}).get_text()
    return stars.strip()
  except Exception as e:
    print(e)


# getting the title, but sometimes returns 'NoneType' object has no attribute 'get_text', need to fix
def get_title(a_soup):
  try:
    title = a_soup.find("a", {"class": "a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold"}).get_text()
    return title.strip()
  except Exception as e:
    pass                          # passing error for now
    #print(e)


# getting the date
def get_date(a_soup):
  try:
    date = a_soup.find("span", {"class": "review-date"}).get_text()
    return date.strip()
  except Exception as e:
    print(e)


# getting the main text of the review
def get_body(a_soup):
  try:
    body = a_soup.find("span", {"class": "a-size-base review-text review-text-content"}).get_text()
    return body.strip()
  except Exception as e:
    print(e)


# function to put all of this together in a dictionary to return names of the columns as keys and data of each review as values
def get_all_components(a_review):             
  return {
      "Username": get_usernames(a_review),
      "Stars": get_stars(a_review),
      "Title": get_title(a_review),
      "Date": get_date(a_review),
      "Review": get_body(a_review)
  }



# function to request the html of any page
def get_page_html(page_url):
    resp = requests.get(page_url, headers=headers)   # sending request with the headers
    return resp.text                                 # returning response as html


# function to find reviews from any page
def get_reviews_from_html(page_html):
    soup = BeautifulSoup(page_html, "lxml")          # parsing html with BeautifulSoup, using the lxml parser
    reviews = soup.find_all("div", {"class": "a-section celwidget"}) # finding the section where each review is (all components)
    return reviews                                                   # returning them



# appending all data to a list
results = []                               # creating a new list
for u in urls_us:                           
  html = get_page_html(u)                  # checking all url pages as html
  reviews = get_reviews_from_html(html)    # getting reviews for each url
  for r in reviews:
    data = get_all_components(r)         # getting the data for all 5 fields considered for each review
    results.append(data)                   # appending the data to the list


# doing the same for the UK urls
for u in urls_uk:
  html = get_page_html(u)
  reviews = get_reviews_from_html(html)
  for r in reviews:
    data = get_all_components(r)
    results.append(data)



df1 = pd.DataFrame.from_records(results) # transforming output to a dataframe, using from_records since it is a list of dictionaries
file_name = f"{datetime.now().strftime('%Y-%m-%d-%H:%M')}.csv" #naming file as the date (GMT) using f string and datetime
df1.to_csv(file_name) #transforming and saving dataframe to csv file
```
